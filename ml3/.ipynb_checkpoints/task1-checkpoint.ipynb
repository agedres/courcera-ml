```python
import numpy as np
from matplotlib import pyplot as plt
from random import randint as rnd

from sklearn.linear_model import LogisticRegression

from PIL import Image
from scipy.io import loadmat
```

```python
#displays random k*k digits from dataset
def show_data(X, k=10):
    m, n = np.shape(X)
    
    mat = X[rnd(0, m - 1),:].reshape((20, 20)).T
    for _ in range(k - 1):
        mat = np.append(mat, X[rnd(0, m - 1),:].reshape((20, 20)).T, axis=1)
    
    for _ in range(k - 1):
        tmp = X[rnd(0, m - 1),:].reshape((20, 20)).T
        for _ in range(k - 1):
            tmp = np.append(tmp, X[rnd(0, m - 1),:].reshape((20, 20)).T, axis=1)
        
        mat = np.append(mat, tmp, axis=0)
    
    im = Image.fromarray(mat * 255)
    imr = im.resize((500, 500), Image.ANTIALIAS)
    plt.imshow(np.asarray(imr), cmap='Greys_r')
    plt.show()
```

```python
#load data
data = loadmat('mnist.mat')
X = data['X']
y = data['y']
```

```python
#show 100 digits in grid
show_data(X, k=10)
```

```python
#train and evaluate model
yt = y.reshape(np.shape(y)[0])

model = LogisticRegression()
model.fit(X, yt)

score = model.score(X, yt) * 100
print("Model accuracy = {:3.2f}%".format(score))
```

```python
#try make predictions for random examples from dataset
m, n = np.shape(X)

for _ in range(10):
    r = rnd(0, m - 1)
    ans = model.predict(X[r,:].reshape(1, -1))[0]
    print("Prediction for {}-th ex. from dataset is: {}. Actual class is: {}".format(r, ans, yt[r]))
```
